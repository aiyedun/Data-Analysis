## 1. DATA VISUALISATION

Data visualisation - summarises information into graphical form. Think of an entire excel sheet with 10000 rows being summarised into a bar chart!

1. Accuracy - data analysts should prioritise accuracy of the data at all times. From cleaning and standardising the data, the data should not be skewed in anyway

2. Helpful - the exploratory process should inform the audience. Afterall the whole idea is to expose patterns within the data

3. Scalability - the visualisations should meet the needs of all stakeholders irrespective of their depth ( with regards to technicalities).

## 1.1 Bar Chart


A bar plot is a graph that represents categorical data with rectangular lengths and heights that are proportional to the values they represent.

## 1.2 `Histogram`
Histogram groups data into bins of equal width. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin.
With a histogram, data is grouped into ranges and each range corresponds to a vertical bar. Histogram is mainly used when the user is considering a continous stream of data.  

A continous data is a stream of data that has several values. An example is the ages of all the children in Ghana's 31 plus million population


image.png

## 1.3 Scatter Plots

Scatter plots use dots to represent values for two different numeric parameters. With scatter plots, the position of each dot on the horizontal and vertical axis indicates values for an individual data point.

**Example**: Comparing the weight and height of members of Pydata beginner bootcamp. Every dot represents a data point, a member. With these parameters represented on the grid, when we consider the data as a whole, we are able to draw insights from it. How many of the members fall within the average belt, are there outliers at the extreme ends? among other insights.


image.png


## 1.4 Box Plot

Box plot/ whisker plot shows the distribution of data and skewness by dividing a numerical dataset into quartiles. Box plots divide a dataset into a range of 25% and giving the position of the middle sections and outliers, users are able to make conclusions on the nature of the data under scrutiny. Represented by a rectangular box with whiskers on either sides of the box, the 5-member summary given by a box plot includes:
      - Minimum score
      - first quartile
      - second quartile
      - median
      - maximum score

````python
# add grids to the graph
# Line plot with two lines

#data
x = [1,2,3]
y = [2,1,4]
x2 = [4, 2, 1]

# x-label
plt.xlabel('X- axis')

# y-label
plt.ylabel('Y- axis')

# title
plt.title('Two Lines Now')

# plotting first line
plt.plot(x,y, label='Line1')

# plotting second line
plt.plot(x, x2, label='Line2')

# adds a legend to the graph
plt.legend( loc= 'best' )

# includes a grid
plt.grid(axis='y')

plt.show()
````

````python
# Line plot with two lines
x = [1,2,3]
y = [2,1,4]
x2 = [4, 2, 1]

plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.title('Two Lines Now')


# ls = ':'
plt.plot(x,y, label='Line1', color ='#4CAF50', linestyle='dashed', marker='>')

# ms represents marker size
plt.plot(x, x2, label='Line2', color ='b', linestyle='dotted', marker='^', ms=20) #ms=20

# adds a legend to the graph
plt.legend( loc= 'best' )

plt.show()


````



1.  The linestyle paramter changes the style of the plotted line.
 You can specify **linestyles**  or **ls** by using strings like 'dotted'/ (:), 'dashed'/ (--), 'solid' etc but the default, linestyle is 'solid'
2.   The marker emphasises a datapoint. There are filled and unfilled **markers** and so depending on the user preference there is a variety of markers to choose from. Using the parameter **ms**, you can specify the size of the markers
3. The color parameter affects the colour of the line.
There are built-in colors that have been simplified within matplotlib, beyond these colors, you would have to specify their hexadecimal color codes. eg. '#4CAF50'

## ** Exploratory Data Analysis (EDA)

With exploratory data analysis, we describe data by means of statistical and visualisation techniques. The purpose of EDA is to bring the important facets of a dataset into focus.


Let's remind ourselves of a few concepts in pandas:
Working with pandas requires two main data structures, the **Series** and the __Dataframe__
1. Series - think of the Series as a column on a spreadsheet. it is a 1D array which houses various datatypes (int, float, etc). The values in the Series have positions / labels called **index**

2. Dataframe - a 2D array with both rows and columns. A typical table that contains information. Similar to numpy, pandas has:

- .shape - gives the number of rows and columns present in a dataframe
-  .size which work in the same way as in numpy

You can fall on these functions to take a quick look at your dataset:
- .head() - by default prints the first 5 rows of the dataset

- .tail() - defaults to printing the last 5 rows of the dataset

- .info() - gives an overview of the data. the null-values, index, column names, datatypes and memory usage

- .describe() - gives a statistical summary of the dataset

- .loc - allows us to section a given dataset by rows and columns using their labels or a conditional statement

- .iloc - when you do know the index of the section you want, you use iloc

- .columns - prints the column names
